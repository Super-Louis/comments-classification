```python
import json
import jieba
import pandas as pd
import pyprind
import re
file = open('../data/jd-phones.json')
df = pd.DataFrame()
pbar = pyprind.ProgBar(441295)
for l in file:
    content = json.loads(l)
    score = content['comment_detail']['score']
    comment = content['comment_detail']['content']
    comment = re.sub('[\W]+', ' ', comment)
    if not (score or comment):
        continue
    if comment[:3] == '此用户':
        continue
    sentence = jieba.cut(comment)
    sentence = ' '.join(sentence)
    print(sentence, score)
    label = 0 if score <= 3 else 1 # 小于等于3为反例（负面）
    pbar.update()
    df = df.append([[sentence, label]], ignore_index=True)
```
