```python
import keras
from keras.models import load_model
```

```python
model = load_model('lstm_keras')
```

```python
import jieba
import json
```

```python
with open('../data/word2num', 'r') as f:
    word2num = json.load(f)
```

```python
comments = '耳机插进去没有用，还是一直外放音乐，有点失望哦，请问下啷个解决呢?'
comments_list = jieba.lcut(comments)
```

```python
x = [word2num[w] for w in comments_list if w in word2num]
```

```python
x
```

```python
import numpy as np
```

```python
result = model.predict(np.array([x]), batch_size=32) # 输入一定要是一个numpy array!
```

```python
result[0,0]
```

```python
def predict_category(sentence):
    wl = jieba.lcut(sentence)
    x_ = [word2num[w] for w in wl if w in word2num]
    x = np.array([x_])
    rs = model.predict(x, batch_size=1)
    print("predict result: {}".format(rs))
    if rs < 0.5:
        return 0
    else:
        return 1
```

```python
sen = '手机用了一天才来评论，玩王者荣耀2个小时，摄像头部位会很烫，而且非常耗电，70%开始游戏，2个小时后只剩15%，新机就这样，不知道用上半年会怎样。好的地方，外观漂亮，拍照历害。音质不错。'
result = predict_category(sen)
print("predict label: {}".format(result))
```
