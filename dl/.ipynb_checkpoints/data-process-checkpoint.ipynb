```python
import pandas as pd
```

```python
df = pd.read_csv('../data/csvfinal.csv')
df['sentence'][2]
```

```python
X = df.iloc[:,0].values
```

```python
Y = df.iloc[:, 1].values
```

```python
X_str = ' '.join(X)
word_dict = {}
for i in X_str.split():
    word_dict[i] = word_dict.get(i, 0) + 1
word_dict
```

```python
word_dict['padded'] = -1
voc = sorted(word_dict.items(), key=lambda x: x[1])
```

```python
filtered_voc = list(filter(lambda x: x[1] > 2, voc))
```

```python
filtered_voc.insert(0, ('padded', -1))
```

```python
stop_words = open('../data/stop_words.txt', 'r', encoding='gbk').readlines()
```

```python
stop_words = [w.strip() for w in stop_words]
```

```python
filtered_voc = list(filter(lambda x: x[0] not in stop_words, filtered_voc))
filtered_voc
```

```python
len(filtered_voc)
```

```python
word2num = {filtered_voc[i][0]: i for i in range(len(filtered_voc))}
```

```python
word2num['pad']
```

```python
num2word = {i : filtered_voc[i][0] for i in range(len(filtered_voc))}
```

```python
X_ = []
for s in X:
    x_ = []
    s = s.split()
    for w in s:
        if w in word2num:
            x_.append(word2num[w])
    X_.append(x_)
```

```python
import numpy as np
```

```python
indexes = np.random.permutation(len(X_))
X_final = np.array(X_)[indexes]
y_final = np.array(Y)[indexes]
```

```python
words = [num2word[n] for n in X_final[1]]
```

```python
words
```

```python
y_final[1]
```

```python
np.save('../data/X_lstm', X_final)
```

```python
np.save('../data/Y_lstm', y_final)
```
